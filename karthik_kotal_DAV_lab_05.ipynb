```python
import numpy as np
import matplotlib.pyplot as plt
games = np.array([1, 2, 3, 4, 5, 6])
attendance = np.array([1500, 1800, 1200, 1600, 2000, 1900])
average_attendance = np.mean(attendance)
print(f"Average attendance over the season: {average_attendance:.2f}")
plt.figure(figsize=(8, 5))
plt.plot(games, attendance, marker='o', color='green', label='Attendance per Game')
plt.axhline(y=average_attendance, color='red', linestyle='--', label=f'Average Attendance')
plt.title("Attendance per Game Over the Season")
plt.xlabel("Game Number")
plt.ylabel("Attendance")
plt.xticks(games)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
players = ['A', 'B', 'C', 'D', 'E']
points = np.array([
    [20, 15, 25, 18, 17],
    [22, 18, 20, 25, 17],
    [15, 12, 28, 10, 13],
    [18, 20, 22, 15, 13],
    [25, 25, 30, 20, 10],
    [20, 22, 18, 30, 10]
])
total_points = np.sum(points, axis=0)
top_player_index = np.argmax(total_points)
top_player = players[top_player_index]
top_score = total_points[top_player_index]
print("Total points scored by each player:")
for p, s in zip(players, total_points):
    print(f"{p}: {s}")
print(f"\nPlayer {top_player} scored the most points: {top_score}")
plt.figure(figsize=(8, 5))
bars = plt.bar(players, total_points, color='skyblue')
bars[top_player_index].set_color('orange')
plt.title("Total Points Scored by Each Player")
plt.xlabel("Player")
plt.ylabel("Total Points Scored")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend([bars[top_player_index]], [f'Top Scorer: {top_player} ({top_score})'])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
points_scored = np.array([95, 102, 78, 88, 110, 100])
threshold = 100
games_above_threshold = np.sum(points_scored > threshold)
print(f"Number of games scoring above {threshold} points: {games_above_threshold}")
bins = [70, 80, 90, 100, 110, 120]
counts, bin_edges = np.histogram(points_scored, bins=bins)
labels = [f"{int(bin_edges[i])}-{int(bin_edges[i+1])}" for i in range(len(bin_edges)-1)]
plt.figure(figsize=(8, 5))
plt.bar(labels, counts, color='black')
plt.title("Number of Games Scored in Different Point Ranges")
plt.xlabel("Points Scored Range")
plt.ylabel("Number of Games")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
games = np.array([1, 2, 3, 4, 5, 6])
opponents = np.array(['Team A', 'Team B', 'Team C', 'Team D', 'Team E', 'Team F'])
points_scored = np.array([95, 90, 110, 95, 100, 105])  # Example scored points
opponent_points = np.array([90, 88, 100, 85, 95, 100])
point_diff = points_scored - opponent_points
best_index = np.argmax(point_diff)
best_opponent = opponents[best_index]
best_diff = point_diff[best_index]
print("Points scored vs opponents:")
for o, ps, op, diff in zip(opponents, points_scored, opponent_points, point_diff):
    print(f"{o}: Scored {ps}, Opponent {op}, Difference {diff:+}")
print(f"\nBest performance: Against {best_opponent} (won by {best_diff} points)")
plt.figure(figsize=(10,6))
bars = plt.bar(opponents, points_scored, color='skyblue')
bars[best_index].set_color('orange')
plt.title("Team's Points Scored Against Each Opponent")
plt.xlabel("Opponent")
plt.ylabel("Points Scored")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend([bars[best_index]], [f'Best vs {best_opponent} (+{best_diff})'])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Game": [1, 2, 3, 4, 5, 6],
    "PointsScored": [95, 102, 78, 88, 110, 100],
    "Opponent Points": [90, 88, 100, 85, 95, 100],
    "Attendance": [1500, 1800, 1200, 1600, 2000, 1900],
    "Opponent": ["A", "B", "C", "D", "E", "F"]
}

df = pd.DataFrame(data)

# Determine win/loss/tie
df["Result"] = np.where(df["PointsScored"] > df["Opponent Points"], "Win",
                 np.where(df["PointsScored"] < df["Opponent Points"], "Loss", "Tie"))

# Calculate counts and average points scored per result
summary = df.groupby("Result").agg(count=("Result", "size"),
                                   mean_points=("PointsScored", "mean")).reset_index()

order = ["Win", "Loss", "Tie"]
summary["Result"] = pd.Categorical(summary["Result"], categories=order, ordered=True)
summary = summary.sort_values("Result")

x = np.arange(len(summary))
width = 0.35

fig, ax = plt.subplots(figsize=(8,5))

bars1 = ax.bar(x - width/2, summary["count"], width, label="Games (Count)", color='skyblue')
bars2 = ax.bar(x + width/2, summary["mean_points"], width, label="Avg Points Scored", color='orange')

ax.set_title("Win/Loss Record vs. Average Points Scored")
ax.set_xlabel("Result Type")
ax.set_ylabel("Count / Average Points")
ax.set_xticks(x)
ax.set_xticklabels(summary["Result"])
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1 + bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.5, f"{height:.1f}", ha='center')

plt.tight_layout()
plt.show()
```
